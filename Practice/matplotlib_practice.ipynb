```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
y_axis = [10.02, 23.24, 39.2, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='green', marker='d', linewidth=2, label='Boston')
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.grid()
```

```python
plt.bar(x_axis, y_axis)
```

```python
plt.bar(x_axis, y_axis, color='green', label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.title('PyBer Fare per Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.legend()
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k', label='New York')
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='c', label='Chicago')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
plt.gca().invert_yaxis()
ax.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='r', label='Chicago')
plt.ylabel('Date')
plt.xlabel('Fare ($)')
plt.xlim(0, 45)
plt.gca().invert_yaxis()
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [10 * i for i in y_axis], c='skyblue', alpha=0.8, edgecolor='k', linewidth=2, label='Boston')
ax.set_title('PyBer Fare by Date')
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
```

```python
plt.subplots(figsize=(8,8))
explode = [0,0,0,0,0,0,0.2,0,0,0,0,0]
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']
plt.pie(y_axis, labels=x_axis, explode=explode, autopct='%1.1f%%', colors=colors)
```

```python
explode[2] = 0.2
fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, labels=x_axis, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, counterclock=False, startangle=90+(y_axis[0]/sum(y_axis)*360/2))
```
